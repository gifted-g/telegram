```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.linkedin.com/jobs/search/?currentJobId=4106957541&f_C=75752648&geoId=92000000&origin=COMPANY_PAGE_JOBS_CLUSTER_EXPANSION&originToLandingJobPostings=4106957541"
page = requests.get(url)
soup = BeautifulSoup(page.text,'html')
print(soup)
```

```python
soup.fin_all('div', class_ = 'base-card relative w-full hover:no-underline focus:no-underline base-card--link base-search-card base-search-card--link job-search-card')
```
